```python
import numpy as np
import matplotlib.pyplot as plt

# Define the thermal distribution
def p(n, n_bar):
    return (n_bar**n) / ((1 + n_bar)**(n + 1))

# Range of n_bar values
n_bar_values = np.linspace(0, 10, 1000)

# Plot for n = 1, 2, 3
plt.figure(figsize=(8, 6))
for n in [1, 2, 3]:
    plt.plot(n_bar_values, p(n, n_bar_values), label=f'n = {n}')

# Add labels and legend
plt.xlabel(r'$\bar{n}$', fontsize=14)
plt.ylabel(r'$p(n)$', fontsize=14)
plt.title('Thermal Distribution: $p(n)$ vs. $barn$', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la distribucin trmica
def p(n, n_bar):
    return (n_bar**n) / ((1 + n_bar)**(n + 1))

# Rango de valores de n_bar
n_bar_values = np.linspace(0, 10, 1000)

# Crear la figura
plt.figure(figsize=(8, 6))

# Graficar para n = 1, 2, 3
colors = ['blue', 'orange', 'green']
for n in [1, 2, 3]:
    plt.plot(n_bar_values, p(n, n_bar_values), label=f'n = {n}')
    # Aadir una lnea vertical en n_bar = n
    plt.axvline(x=n, color=colors[n-1], linestyle='--', alpha=0.5, label=f'$\\bar{{n}} = {n}$')

# Aadir etiquetas y leyenda
plt.xlabel(r'$\bar{n}$', fontsize=14)
plt.ylabel(r'$p(n)$', fontsize=14)
plt.title(r'Thermal distribution: $p(n)$ vs. $\bar{n}$', fontsize=16)
plt.legend()
plt.grid(True)
plt.show()
```
